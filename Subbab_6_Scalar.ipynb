# SUBBAB 6: SCALAR

**NO SUBBAB : 6**

Topik ini membahas tentang skalar — nilai tunggal tanpa dimensi (rank 0 tensor).

## Soal No. 1 — Python Native

Buat variabel skalar `x` dan simpan suatu nilai skalar.

```python
# Buat variabel skalar x dan simpan suatu nilai skalar
x = 10
x
```

```python
# Tampilkan tipe data dari x
type(x)
```

```python
# Buat variabel skalar y
y = 5
y
```

```python
# Jumlahkan x + y
py_sum = x + y
py_sum
```

```python
# Tampilkan tipe data dari py_sum
type(py_sum)
```

```python
# Buat variabel x_float (desimal) dan jumlahkan dengan y
x_float = 3.14
float_sum = x_float + y
float_sum
```

```python
# Tampilkan tipe data dari float_sum
type(float_sum)
```

## Soal No. 2 — TensorFlow

Ulangi soal No. 1 menggunakan TensorFlow.

```python
import tensorflow as tf
```

```python
# Buat variabel skalar x menggunakan TensorFlow
x_tf = tf.Variable(10, dtype=tf.int16)
x_tf
```

```python
# Tampilkan shape dari x_tf
x_tf.shape
```

```python
# Buat variabel skalar y
y_tf = tf.Variable(5, dtype=tf.int16)
y_tf
```

```python
# Jumlahkan x_tf + y_tf
x_tf + y_tf
```

```python
# Jumlahkan menggunakan tf.add
tf_sum = tf.add(x_tf, y_tf)
tf_sum
```

```python
# Konversi ke numpy
tf_sum.numpy()
```

```python
# Tampilkan tipe data dari tf_sum
type(tf_sum.numpy())
```

```python
# Buat variabel float desimal
x_float_tf = tf.Variable(3.14, dtype=tf.float16)
x_float_tf
```

## Soal No. 3 — PyTorch

Ulangi soal No. 1 menggunakan PyTorch.

```python
import torch
```

```python
# Buat variabel skalar x menggunakan PyTorch
x_pt = torch.tensor(10)
x_pt
```

```python
# Tampilkan shape dari x_pt
x_pt.shape
```

```python
# Buat variabel skalar y
y_pt = torch.tensor(5)
y_pt
```

```python
# Jumlahkan x_pt + y_pt
pt_sum = x_pt + y_pt
pt_sum
```

```python
# Tampilkan tipe data dari pt_sum
type(pt_sum)
```

```python
# Buat variabel float desimal
x_float_pt = torch.tensor(3.14, dtype=torch.float16)
x_float_pt
```

```python
# Tampilkan shape dari x_float_pt
x_float_pt.shape
```
