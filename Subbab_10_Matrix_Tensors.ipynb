# SUBBAB 10: Matrix Tensors

Topik ini membahas tentang **matrix** — tensor 2 dimensi (rank 2).

Matrix berukuran **m x n** artinya memiliki **m baris** dan **n kolom**.

```python
import numpy as np
import torch
import tensorflow as tf
```

## Soal No. 1 — NumPy

Buat matrix X berukuran 2x3 menggunakan NumPy.

```python
# Buat matrix X berukuran 2x3 menggunakan numpy
X = np.array([[1, 2, 3],
              [4, 5, 6]])
print('Matrix X (2x3):')
print(X)
```

```python
# Tampilkan ukuran dari matrix X
print('Ukuran matrix X (shape):', X.shape)
```

```python
# Tampilkan seluruh elemen dari kolom pertama matrix X
kolom_pertama = X[:, 0]
print('Kolom pertama matrix X:', kolom_pertama)
```

```python
# Tampilkan seluruh elemen dari baris kedua matrix X
baris_kedua = X[1, :]
print('Baris kedua matrix X:', baris_kedua)
```

## Soal No. 2 — PyTorch

Ulangi soal nomor 1 menggunakan PyTorch.

```python
# Buat matrix X berukuran 2x3 menggunakan PyTorch
X_pt = torch.tensor([[1, 2, 3],
                     [4, 5, 6]])
print('Matrix X (PyTorch):')
print(X_pt)
```

```python
# Tampilkan ukuran dari matrix X
print('Ukuran matrix X (shape):', X_pt.shape)
```

```python
# Tampilkan seluruh elemen dari kolom pertama
print('Kolom pertama:', X_pt[:, 0])
```

```python
# Tampilkan seluruh elemen dari baris kedua
print('Baris kedua:', X_pt[1, :])
```

## Soal No. 2 (lanjutan) — TensorFlow

Ulangi soal nomor 1 menggunakan TensorFlow.

```python
# Buat matrix X berukuran 2x3 menggunakan TensorFlow
X_tf = tf.constant([[1, 2, 3],
                    [4, 5, 6]])
print('Matrix X (TensorFlow):')
print(X_tf)
```

```python
# Tampilkan ukuran dari matrix X
print('Ukuran matrix X (shape):', X_tf.shape)
```

```python
# Tampilkan seluruh elemen dari kolom pertama
print('Kolom pertama:', X_tf[:, 0])
```

```python
# Tampilkan seluruh elemen dari baris kedua
print('Baris kedua:', X_tf[1, :])
```
