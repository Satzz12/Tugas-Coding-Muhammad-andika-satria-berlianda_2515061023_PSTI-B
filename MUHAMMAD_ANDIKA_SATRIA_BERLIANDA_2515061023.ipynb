SOAL NO. 2 " TENSORFLOW"


```python
import tensorflow as tf
```

```python
# Soal 2 - Buat variabel skalar x dan simpan suatu nilai skalar
x_tf = tf.Variable(10, dtype=tf.int16)
x_tf
```

```python
# Tampilkan tipe data dari x
x_tf.shape
```

```python
# Buat variabel skalar y dan simpan suatu nilai skalar
y_tf = tf.Variable(5, dtype=tf.int16)
```

```python
# Jumlahkan x + y
x_tf + y_tf
```

```python
# Jumlahkan x_tf dan y_tf menggunakan tf.add
tf_sum = tf.add(x_tf, y_tf)
tf_sum
```

```python
tf_sum.numpy()
```

```python
# Tampilkan tipe data dari tf_sum
type(tf_sum.numpy())
```

```python
# Buat variabel bernama x_float dan simpan suatu nilai desimal
x_float = tf.Variable(3.14, dtype=tf.float16)
x_float
```
