# SUBBAB 11: Generic Tensor Notation

Topik ini membahas notasi tensor umum dan cara membaca dimensi tensor.

**Tensor** adalah generalisasi dari skalar, vektor, dan matriks:
- Rank 0 → Skalar (nilai tunggal)
- Rank 1 → Vektor (1D)
- Rank 2 → Matriks (2D)
- Rank 3+ → Tensor tingkat tinggi (3D, 4D, ...)

````python
import torch
import numpy as np
````

## Soal No. 1 — Jelaskan perintah berikut:

```python
images_pt = torch.zeros([32, 28, 28, 3])
```

Perintah ini membuat tensor 4 dimensi berisi semua nilai **0** dengan:
| Dimensi | Nilai | Arti |
|---------|-------|------|
| 0 | 32 | Batch size — jumlah gambar |
| 1 | 28 | Height — tinggi gambar (piksel) |
| 2 | 28 | Width — lebar gambar (piksel) |
| 3 | 3  | Channels — RGB (Merah, Hijau, Biru) |

````python
# Buat tensor images_pt
images_pt = torch.zeros([32, 28, 28, 3])
print('Tensor images_pt dibuat!')
````

````python
# Tampilkan shape (ukuran) tensor
print('Shape:', images_pt.shape)
````

````python
# Tampilkan dtype (tipe data) tensor
print('Dtype:', images_pt.dtype)
````

````python
# Tampilkan jumlah dimensi (rank) tensor
print('Jumlah dimensi (ndim):', images_pt.ndim)
````

````python
# Tampilkan total elemen tensor
print('Total elemen:', images_pt.numel())
````

````python
# Penjelasan rinci setiap dimensi
print('Penjelasan dimensi tensor images_pt:')
print(f'  Dimensi-0 (batch size)  : {images_pt.shape[0]} gambar')
print(f'  Dimensi-1 (height)      : {images_pt.shape[1]} piksel')
print(f'  Dimensi-2 (width)       : {images_pt.shape[2]} piksel')
print(f'  Dimensi-3 (channels)    : {images_pt.shape[3]} (RGB)')
````

````python
# Akses satu gambar dari batch (gambar index ke-0)
satu_gambar = images_pt[0]
print('Shape satu gambar:', satu_gambar.shape)  # Hasilnya [28, 28, 3]
````

````python
# Akses satu baris piksel dari gambar pertama
satu_baris = images_pt[0][0]
print('Shape satu baris piksel:', satu_baris.shape)  # Hasilnya [28, 3]
````

````python
# Akses satu piksel (baris 0, kolom 0) dari gambar pertama
satu_piksel = images_pt[0][0][0]
print('Nilai satu piksel (RGB):', satu_piksel)  # Hasilnya [0., 0., 0.] karena semua zeros
````
