# SUBBAB 7: Vectors and Vector Transposition

Topik ini membahas tentang vektor dan operasi transpose menggunakan NumPy, PyTorch, dan TensorFlow.

```python
import numpy as np
import torch
import tensorflow as tf
```

## Soal No. 1 — NumPy

Buat vector x dengan ukuran 3x1.

```python
# Buat vector x dengan ukuran 3x1
x = np.array([[1],
              [2],
              [3]])
print('Vector x (3x1):')
print(x)
```

```python
# Hitung panjang vector x
panjang_x = len(x)
print('Panjang vector x:', panjang_x)
```

```python
# Tampilkan ukuran vector x
print('Ukuran vector x (shape):', x.shape)
```

```python
# Tampilkan tipe data dari vector x
print('Tipe data vector x:', type(x))
print('Dtype elemen:', x.dtype)
```

```python
# Akses elemen pertama dari vector x
elemen_pertama = x[0]
print('Elemen pertama vector x:', elemen_pertama)
```

```python
# Tampilkan tipe data elemen pertama dari vector x
print('Tipe data elemen pertama:', type(elemen_pertama))
```

## Soal No. 2 — Transpose Vector (NumPy)

Lakukan transpose pada vector x dan simpan pada variabel `x_t`.

```python
# Lakukan transpose pada vector x dan simpan pada variabel x_t
x_t = x.T
print('Transpose x_t:')
print(x_t)
```

```python
# Tampilkan ukuran dari vector x_t
print('Ukuran x_t:', x_t.shape)
```

```python
# Buat vector y dan lakukan transpose pada vector y, simpan pada variabel y_t
y = np.array([[4],
              [5],
              [6]])
y_t = y.T
print('Transpose y_t:')
print(y_t)
```

```python
# Tampilkan ukuran dari vector y_t
print('Ukuran y_t:', y_t.shape)
```

## Soal No. 3 — Transpose dengan PyTorch dan TensorFlow

Lakukan soal nomor 2 menggunakan PyTorch dan TensorFlow.

```python
# ======== PyTorch ========
x_pt = torch.tensor([[1.0], [2.0], [3.0]])
x_t_pt = x_pt.T
print('Vector x (PyTorch):')
print(x_pt)
print('Transpose x_t (PyTorch):')
print(x_t_pt)
print('Ukuran x_t PyTorch:', x_t_pt.shape)
```

```python
y_pt = torch.tensor([[4.0], [5.0], [6.0]])
y_t_pt = y_pt.T
print('Transpose y_t (PyTorch):')
print(y_t_pt)
print('Ukuran y_t PyTorch:', y_t_pt.shape)
```

```python
# ======== TensorFlow ========
x_tf = tf.constant([[1.0], [2.0], [3.0]])
x_t_tf = tf.transpose(x_tf)
print('Vector x (TensorFlow):')
print(x_tf)
print('Transpose x_t (TensorFlow):')
print(x_t_tf)
print('Ukuran x_t TensorFlow:', x_t_tf.shape)
```

```python
y_tf = tf.constant([[4.0], [5.0], [6.0]])
y_t_tf = tf.transpose(y_tf)
print('Transpose y_t (TensorFlow):')
print(y_t_tf)
print('Ukuran y_t TensorFlow:', y_t_tf.shape)
```
