# SUBBAB 9: Basis, Orthogonal, and Orthonormal Vectors

Dua vektor dikatakan **orthogonal** jika **dot product**-nya = 0:
$$u \cdot v = 0$$

Dua vektor dikatakan **orthonormal** jika orthogonal DAN masing-masing memiliki panjang = 1.

```python
import numpy as np
```

## Soal No. 1 — Vector i = (1,0) dan vector j = (0,1)

```python
# Buat vector i dengan elemen (1,0) dan vector j dengan elemen (0,1)
i = np.array([1, 0])
j = np.array([0, 1])
print('Vector i:', i)
print('Vector j:', j)
```

```python
# Buktikan bahwa vector i dan vector j itu orthogonal
dot_ij = np.dot(i, j)
print('Dot product i . j:', dot_ij)
if dot_ij == 0:
    print('✓ Vector i dan j ORTHOGONAL (dot product = 0)')
else:
    print('✗ Vector i dan j TIDAK orthogonal')
```

```python
# Cek apakah i dan j orthonormal (norm masing-masing harus = 1)
norm_i = np.linalg.norm(i)
norm_j = np.linalg.norm(j)
print('Norm vector i:', norm_i)
print('Norm vector j:', norm_j)
if np.isclose(norm_i, 1.0) and np.isclose(norm_j, 1.0) and dot_ij == 0:
    print('✓ Vector i dan j adalah ORTHONORMAL')
else:
    print('✗ Vector i dan j bukan orthonormal')
```

## Soal No. 2 — v = (1,2) dan w = (2,-1), apakah orthogonal?

```python
# Buat vector v = (1,2) dan w = (2,-1)
v = np.array([1, 2])
w = np.array([2, -1])
print('Vector v:', v)
print('Vector w:', w)
```

```python
# Hitung dot product v . w
dot_vw = np.dot(v, w)
print('Dot product v . w:', dot_vw)
if dot_vw == 0:
    print('✓ Vector v dan w ORTHOGONAL (dot product = 0)')
else:
    print('✗ Vector v dan w TIDAK orthogonal')
```

```python
# Cek apakah v dan w orthonormal
norm_v = np.linalg.norm(v)
norm_w = np.linalg.norm(w)
print('Norm vector v:', norm_v)
print('Norm vector w:', norm_w)
if np.isclose(norm_v, 1.0) and np.isclose(norm_w, 1.0) and dot_vw == 0:
    print('✓ Vector v dan w adalah ORTHONORMAL')
elif dot_vw == 0:
    print('⚠ Vector v dan w ORTHOGONAL tapi BUKAN orthonormal (norm bukan 1)')
else:
    print('✗ Bukan orthogonal maupun orthonormal')
```
