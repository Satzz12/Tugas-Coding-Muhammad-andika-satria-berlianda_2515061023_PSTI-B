# SUBBAB 8: Norms and Unit Vectors

Topik ini membahas tentang **L2 Norm** (Euclidean norm) dan **Unit Vector**.

- **L2 Norm**: $\|x\| = \sqrt{\sum x_i^2}$
- **Unit Vector**: $\hat{v} = \dfrac{v}{\|v\|}$, memiliki panjang = 1

```python
import numpy as np
```

## Soal No. 1 — L2 Norm Vector x (3x1)

```python
# Buat vector x dengan ukuran 3x1
x = np.array([[1],
              [2],
              [3]])
print('Vector x:')
print(x)
```

```python
# Hitung nilai L2 norm dari vector x
l2_norm_x = np.linalg.norm(x)
print('L2 Norm dari vector x:', l2_norm_x)
```

## Soal No. 2 — Vector v = (3, 4)

```python
# Buat vector v dengan elemen (3, 4)
v = np.array([3, 4])
print('Vector v:', v)
```

```python
# Hitung nilai L2 norm dari vector v
l2_norm_v = np.linalg.norm(v)
print('L2 Norm dari vector v:', l2_norm_v)
```

```python
# Tampilkan unit vector dari vector v
# Unit vector = v / ||v||
unit_v = v / l2_norm_v
print('Unit vector dari v:', unit_v)
```

```python
# Hitung panjang unit vector dari vector v (harus = 1)
panjang_unit_v = np.linalg.norm(unit_v)
print('Panjang unit vector dari v:', panjang_unit_v)
print('Apakah panjangnya = 1?', np.isclose(panjang_unit_v, 1.0))
```
