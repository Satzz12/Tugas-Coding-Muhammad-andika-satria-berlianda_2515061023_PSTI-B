## Soal No. 1 — What is the transpose of this vector?

Diketahui vector kolom:
$$v = \begin{bmatrix} 25 \\ 2 \\ -3 \\ -25 \end{bmatrix}$$

Tentukan transpos dari vector tersebut!

### Jawaban — NumPy

```python
# Definisikan vector v sebagai column vector (4x1)
v = np.array([[25],
              [2],
              [-3],
              [-25]])
print('Vector v (4x1):')
print(v)
print('Shape v:', v.shape)
```

```python
# Transpose vector v -> menjadi row vector (1x4)
v_T = v.T
print('Transpose v^T (1x4):')
print(v_T)
print('Shape v^T:', v_T.shape)
```

### Jawaban — PyTorch

```python
# Definisikan vector v menggunakan PyTorch
v_pt = torch.tensor([[25],
                     [2],
                     [-3],
                     [-25]], dtype=torch.float32)
print('Vector v (PyTorch):')
print(v_pt)
print('Shape v:', v_pt.shape)
```

```python
# Transpose dengan PyTorch
v_T_pt = v_pt.T
print('Transpose v^T (PyTorch):')
print(v_T_pt)
print('Shape v^T:', v_T_pt.shape)
```

### Jawaban — TensorFlow

```python
# Definisikan vector v menggunakan TensorFlow
v_tf = tf.constant([[25],
                    [2],
                    [-3],
                    [-25]], dtype=tf.float32)
print('Vector v (TensorFlow):')
print(v_tf)
print('Shape v:', v_tf.shape)
```

```python
# Transpose dengan TensorFlow
v_T_tf = tf.transpose(v_tf)
print('Transpose v^T (TensorFlow):')
print(v_T_tf)
print('Shape v^T:', v_T_tf.shape)
```

### Kesimpulan

Transpose dari vector kolom $v$ berukuran **(4×1)** adalah vector baris $v^T$ berukuran **(1×4)**:

$$v^T = \begin{bmatrix} 25 & 2 & -3 & -25 \end{bmatrix}$$
